```python
def read():
    ##assumes dataset in folder
    wines   = pd.read_csv("XWines_Full_100K_wines.csv", low_memory=False, encoding="utf-8", memory_map=True)
    ratings = pd.read_csv("XWines_Full_21M_ratings.csv", low_memory=False, encoding="utf-8", memory_map=True)
    return wines,ratings
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python


# Convert categorical features (Wine Type and Region) to numerical values
wines['Type'] = wines['Type'].astype('category').cat.codes  # Encode wine type
wines['Country'] = wines['Country'].astype('category').cat.codes        # Encode region
wines['Body'] = wines['Body'].astype('category').cat.codes
# Check the transformed DataFrame
print(wines.head())



# Check the updated DataFrame
# print(wines.head())
```

```python
# Split the data into features (X) and target (y)
X = wines[['Type', 'Country', 'Body']]
y = wines['WineID']  # Target is the 'Wine Type'

# Split into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(y.value_counts())
```

```python
# Initialize the decision tree classifier
clf = DecisionTreeClassifier()

# Train the model on the training data
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

skf = StratifiedKFold(n_splits=1)

cv_scores = cross_val_score(clf, X, y, cv=skf)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Score: {cv_scores.mean() * 100:.2f}%")

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy * 100:.2f}%")
```
